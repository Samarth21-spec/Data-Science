```python
import requests
from bs4 import BeautifulSoup
URL="https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off"
page=requests.get(URL)
print(page.text)
```

```python
soup=BeautifulSoup(page.text,'html.parser')
```

```python
result=soup.find_all(class_ ="_4rR01T")
for i in range(0,len(result)):
    result[i]=result[i].text
result
```

```python
result2=soup.find_all(class_="_3LWZlK")
for i in range(0,len(result)):
    result2[i]=result2[i].text
result2[:24]
```

```python
import pandas as pd
d={"mobilename":result,"price":result,"Rating:":result2[:24]}
s=pd.DataFrame(d)
s
```
